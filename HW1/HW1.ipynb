# 1.東區總班級數

```python
df1 = df[df['區域']=='東區'] # 篩選出區域為東區的資料

print(f"東區總班級共 {df1['總班級數'].sum()} 個班") # 計算東區各學校的總班級數的總和，然後印出
```

```python
import matplotlib.pyplot as plt

# 篩選出東區的資料
df_east = df[df['區域'] == '東區']

# 計算東區各學校的總班級數
class_counts = df_east.groupby('學校名稱')['總班級數'].sum()

# 繪製柱狀圖
plt.bar(class_counts.index, class_counts.values, color='skyblue')
plt.title('東區學校總班級數')
plt.xlabel('學校名稱')
plt.ylabel('總班級數')
plt.xticks(rotation=45, ha='right')

plt.show()
```

# 2.北屯區各學校八年級班級數共有？班

```python
df2 = df[df['區域']=='北屯區']   # 篩選出區域為北屯區的資料

degree_list = list(df2['八年級班級數'].unique())   # 獲取北屯區各學校的八年級班級數列表
name_list = list(df2['學校名稱'].unique())      # 獲取北屯區各學校名稱列表
print('北屯區中，')
for degree, name in zip(degree_list, name_list):
    print(f"學校名稱是 {name} 的八年級班級數為 {degree} ")  # loop 輸出學校名稱與八年級班級數

```

```python
import matplotlib.pyplot as plt

# 篩選出北屯區的資料
df_bt = df[df['區域'] == '北屯區']

# 計算各學校八年級班級數的總和
class_counts = df_bt.groupby('學校名稱')['八年級班級數'].sum()

# 繪製長條圖
plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('北屯區各學校八年級班級數')
plt.xlabel('學校名稱')
plt.ylabel('八年級班級數')
plt.xticks(rotation=45, ha='right')

plt.show()
```

# 5.外埔國中的七、八、九年級人數比為多少?

```python
df5_1 = df[df['學校名稱']=='外埔國中']    # 篩選出學校名稱為外埔國中的資料

# 計算外埔國中七、八、九年級男生和女生的總人數
seven=df5_1[['七年級男生數','七年級女生數']].sum().sum()
eight=df5_1[['八年級男生數','八年級女生數']].sum().sum()
nine=df5_1[['九年級男生數','九年級女生數']].sum().sum()

# 計算外埔國中總學生數(一整橫排用axis=1，直排用axis=0)
df5_1['外埔國中總學生數']=df5_1[['七年級男生數','八年級男生數','九年級男生數','七年級女生數','八年級女生數','九年級女生數']].sum(axis=1)
df5_3=df5_1['外埔國中總學生數'].sum()

# 求比率
print(f"七年級的人數占比為: {seven/df5_3}")
print(f"八年級的人數占比為: {eight/df5_3}")
print(f"九年級的人數占比為: {nine/df5_3}")
```

```python
import matplotlib.pyplot as plt

# 篩選出外埔國中的資料
df_waipu = df[df['學校名稱'] == '外埔國中']

# 計算各年級的總人數
grade_seven_total = df_waipu['七年級男生數'].sum() + df_waipu['七年級女生數'].sum()
grade_eight_total = df_waipu['八年級男生數'].sum() + df_waipu['八年級女生數'].sum()
grade_nine_total = df_waipu['九年級男生數'].sum() + df_waipu['九年級女生數'].sum()

# 繪製長條圖
plt.figure(figsize=(8, 6))
grades = ['七年級', '八年級', '九年級']
totals = [grade_seven_total, grade_eight_total, grade_nine_total]
plt.bar(grades, totals, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('外埔國中七、八、九年級人數比')
plt.xlabel('年級')
plt.ylabel('人數')
plt.show()
```

# 6.最多人的學校與最少人的學校?


```python
#axis=1 意思是在列方向上執行；axis=0 意思是在行方向上執行

# 計算每所學校的總學生數，並新增一列到df中
df['總學生數']=df[['七年級男生數','八年級男生數','九年級男生數','七年級女生數','八年級女生數','九年級女生數']].sum(axis=1)

# 根據總學生數排序
df6=df.sort_values(by='總學生數', ascending=False)

print(f"最多人的學校是 {df6.head(1)['學校名稱'].iloc[0]}，共有 {df6.head(1)['總學生數'].iloc[0]} 人")
print(f"最少人的學校是 {df6.tail(1)['學校名稱'].iloc[0]}，共有 {df6.tail(1)['總學生數'].iloc[0]} 人")
###

#df6_2=df[['七年級女生數','八年級女生數','九年級女生數']].sum(axis=1)
#df['總學生數']=df6_1+df6_2
###
```

```python
import matplotlib.pyplot as plt

# 計算每個學校的總學生數
df['總學生數'] = df[['七年級男生數', '八年級男生數', '九年級男生數', '七年級女生數', '八年級女生數', '九年級女生數']].sum(axis=1)

# 找出最多人的學校和最少人的學校
max_school = df.loc[df['總學生數'].idxmax()]
min_school = df.loc[df['總學生數'].idxmin()]

print(f"最多人的學校是 {max_school['學校名稱']}，共有 {max_school['總學生數']} 人")
print(f"最少人的學校是 {min_school['學校名稱']}，共有 {min_school['總學生數']} 人")

# 繪製長條圖
plt.figure(figsize=(8, 6))
plt.bar([max_school['學校名稱'], min_school['學校名稱']], [max_school['總學生數'], min_school['總學生數']], color=['skyblue', 'orange'])
plt.title('最多人與最少人的學校總學生數')
plt.xlabel('學校名稱')
plt.ylabel('總學生數')
plt.xticks(rotation=45)
plt.show()
```
